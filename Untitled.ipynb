### we can see that our photos are of different sizes now we will standardize our photo size and bring them to same size.

```python
# Resizing all images 
from keras.preprocessing.image import load_img,img_to_array
from os import listdir
from numpy import asarray
from tqdm import tqdm

photos=[]
labels=[]

folder1='cats/'
for file in tqdm(listdir(folder1)):
    print(file)
    print(folder1+file)
    label=1
    photo=load_img(folder1+file , target_size=(200,200))
    photo=img_to_array(photo)
    photos.append(photo)
    labels.append(label)
 

    
```

```python

folder2='dogs/'
for files in tqdm(listdir(folder2)):
    label=0
    photo=load_img(folder2+files, target_size=(200,200))
    photo=img_to_array(photo)
    photos.append(photo)
    labels.append(label)
    

photos=asarray(photos)
labels=asarray(labels)

print(photos.shape)
print(labels.shape)
```

```python
from numpy import save
# save the reshaped photos
save('dogs_vs_cats_photos.npy', photos)
save('dogs_vs_cats_labels.npy', labels)
```

### we will be creating directory of following structure

```python
dataset_dogs_vs_cats
 test
    cats
    dogs
 train
     cats
     dogs
```

```python
#ref :- https://machinelearningmastery.com/how-to-develop-a-convolutional-neural-network-to-classify-photos-of-dogs-and-cats/
# making directories for train and test data
from os import makedirs
dataset_home = 'dataset_dogs_vs_cats/'
subdirs = ['train/', 'test/']
for subdir in subdirs:
    # create label subdirectories
    labeldirs = ['dogs/', 'cats/']
    for labldir in labeldirs:
        newdir = dataset_home + subdir + labldir
        makedirs(newdir, exist_ok=True)
```

```python
#ref:-https://machinelearningmastery.com/how-to-develop-a-convolutional-neural-network-to-classify-photos-of-dogs-and-cats/
from random import seed
from random import random
from shutil import copyfile
# seed random number generator
seed(1)
# define ratio of pictures to use for validation
val_ratio = 0.25
# copy training dataset images into subdirectories
src_directory = 'cats/'
for file in listdir(src_directory):
    src = src_directory + '/' + file
    dst_dir = 'train/'
    if random() < val_ratio:
        dst_dir = 'test/'
    dst = dataset_home + dst_dir + 'cats/'  + file
    copyfile(src, dst)
src_directory = 'dogs/'
for file in listdir(src_directory):
    src = src_directory + '/' + file
    dst_dir = 'train/'
    if random() < val_ratio:
        dst_dir = 'test/'
    dst = dataset_home + dst_dir + 'dogs/'  + file
    copyfile(src, dst)   
```

### Data Augmentation

```python
from keras.preprocessing.image import ImageDataGenerator

training_data_generator = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True)
```

```python
train_it = training_data_generator.flow_from_directory('dataset_dogs_vs_cats/train/',class_mode='binary', batch_size=64, target_size=(200, 200))
test_it = training_data_generator.flow_from_directory('dataset_dogs_vs_cats/test/',class_mode='binary', batch_size=64, target_size=(200, 200))
```

# Building a CNN model

```python
import keras
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Dense,Flatten,Dropout

model=Sequential()
model.add(Conv2D(32,(3,3),activation='relu',kernel_initializer='he_uniform',input_shape=(200,200,3)))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(64,(3,3),activation='relu',kernel_initializer='he_uniform'))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(128,(3,3),activation='relu',kernel_initializer='he_uniform'))
model.add(MaxPooling2D((2,2)))
model.add(Dropout(.2))
model.add(Flatten())
model.add(Dense(128,activation='relu',kernel_initializer='he_uniform'))
model.add(Dropout(.5))
model.add(Dense(1,activation='sigmoid'))
          
# compiling model
model.compile(optimizer=keras.optimizers.Adam(),loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# fitting model
history = model.fit_generator(train_it, steps_per_epoch=len(train_it),validation_data=test_it, validation_steps=len(test_it), epochs=20, verbose=1)

# evaluating model
_, acc = model.evaluate_generator(test_it, steps=len(test_it), verbose=0)

print('> %.3f' % (acc * 100.0))
```

###  Plotting accuracy and loss

```python
plt.subplot(211)
plt.title('Cross Entropy Loss')
plt.plot(history.history['loss'], color='blue', label='train')
plt.plot(history.history['val_loss'], color='orange', label='test')
plt.legend()
plt.grid()
plt.show()

plt.subplot(212)
plt.title('Classification Accuracy')
plt.plot(history.history['accuracy'], color='blue', label='train')
plt.plot(history.history['val_accuracy'], color='orange', label='test')
plt.legend()
plt.grid()
plt.show()
```

### Making predictions

We will use this below image to predict , If my output is 1 this means my model is predicting Dog and if my output is 0 then my model is predicting cat.

```python
def insert_image(img):
    img=load_img( img ,target_size=(200,200))
    img=img_to_array(img)
    img=img.reshape(1,200,200,3)
    img = img.astype('float32')
    return img
```

```python
image='dog_sample.jpg'
img=insert_image(image)

res=model.predict(img)
print(res)
if res==[[1.]]:
    filename=imread(image)
    plt.imshow(filename)
    plt.title('DOG')
    plt.show()
    
elif res==[[0.]]:
    filename=imread(image)
    plt.imshow(filename)
    plt.title('CAT')
    plt.show()
```
